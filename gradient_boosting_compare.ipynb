```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
```

```python
def confusion_mat(Y_pred, Y_test):
    plt.figure()
    sns.set(font_scale=1.5)
    cm = confusion_matrix(Y_pred, Y_test)
    sns.heatmap(cm, annot=True, fmt='g')
    plt.title('Confusion matrix', y=1.1)
    plt.ylabel('Actual label')
    plt.xlabel('Predicted label')
    plt.show()
```

```python
def compare_metods_on_dataset(df, seed=7, test_size=0.20):
    
    X = df.iloc[:, :-1].values
    Y = df.iloc[:, -1].values
    
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=test_size, random_state=seed)
    
    # Ada Boost -------------
    clf = AdaBoostClassifier(n_estimators=100, random_state=0)
    clf.fit(X_train, y_train)
    predicted = clf.predict(X_test)
    print('AdaBoostClassifier: Accuracy of the result is:')
    print(np.mean(predicted == y_test))
    confusion_mat(predicted, y_test)
    # Gradient Boosting ------------
    clf = GradientBoostingClassifier(n_estimators=100, random_state=0)
    clf.fit(X_train, y_train)
    predicted = clf.predict(X_test)
    print('GradientBoostingClassifier: Accuracy of the result is:')
    print(np.mean(predicted == y_test))
    confusion_mat(predicted, y_test)
    # XGB --------------------
    clf = XGBClassifier(n_estimators=100, random_state=0)
    clf.fit(X_train, y_train)
    predicted = clf.predict(X_test)
    print('XGBClassifier: Accuracy of the result is:')
    print(np.mean(predicted == y_test))
    confusion_mat(predicted, y_test)
    # Cat Boost -------------------
    clf = CatBoostClassifier(iterations=100)
    clf.fit(X_train, y_train)    
    predicted = clf.predict(X_test)
    print('CatBoostClassifier: Accuracy of the result is:')
    print(np.mean(predicted == y_test))
    confusion_mat(predicted, y_test)
```

### VLagun_Phys_Years3.csv

```python
dataset = pd.read_csv('data/VLagun_Phys_Years3.csv')
dataset.head()
```

```python
compare_metods_on_dataset(dataset)
```

#### Wnioski:
Najlepsza metoda dla tego dataset to: AdaBoost, dało najlepsze accuracy

### diabates.csv

```python
dataset = pd.read_csv('data/diabetes.csv')
compare_metods_on_dataset(dataset)
```

#### Wnioski:
Najlepsza metoda dla tego dataset to: CatBoost, dało najlepsze accuracy

### VLagun_Total_Basins_35.csv

```python
df = pd.read_csv('data/VLagun_Total_Basins_35.csv')
df.drop(columns=df.columns[0], axis=1, inplace=True)
compare_metods_on_dataset(df)
```

#### Wnioski:
Najlepsza metoda dla tego dataset to: CatBoost i AdaBoost, dały najlepsze accuracy
